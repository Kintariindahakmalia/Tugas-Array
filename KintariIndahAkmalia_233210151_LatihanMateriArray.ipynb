```python
# Perkalian Matrix

import numpy as np
A = np.array([19,12,24])
B = np.array([86,65,19])
C = np.array([12,23,34])

perkalian = A * B * C
print(perkalian)
```

```python
# Pengurangan Matrix

import numpy as np
A = np.array([19,12,24])
B = np.array([86,65,19])
C = np.array([12,23,34])

pengurangan = A - B - C
print(pengurangan)
```

```python
# Transpose Matrix

import numpy as np
A = np.array([[19,12,24],[86,65,19],[12,23,34]])

transpose_A = np.transpose(A)
print(transpose_A)
```

```python
# Nilai Invers

import numpy as np
A = np.array([[19,12,24],[86,65,19],[12,23,34]])

inverse_A = np.linalg.inv(A)
print("\nInvers matriks A:")
print(inverse_A)
```

```python
# Nilai Determinan

import numpy as np
A = np.array([[19,12,24],[86,65,19],[12,23,34]])

determinant_A = np.linalg.det(A)
print("Determinan dari matriks A:", determinant_A)
```
